```python
# импорт библиотек

import pandas as pd
```

```python
# выгрузка данных из excel-файла (oplogs.xlsx) 

dataframe_1 = pd.read_excel(r"D:\УНИВЕР\Product_analytics\Исходники\oplogs.xlsx")
```

```python
# просматриваем данные

dataframe_1
```

```python
# чистка фрейма от дубликатов

processed_dataframe_1 = dataframe_1.drop_duplicates(keep=False)
```

```python
processed_dataframe_1
```

```python
# узнаём тип данных каждого столбца датафрейма

processed_dataframe_1.dtypes
```

```python
# задаём нужный тип данных для некоторых столбцов и смотрим изменения

processed_dataframe_1["user_id"] = processed_dataframe_1["user_id"] \
                                   .astype("object")
processed_dataframe_1["version"] = processed_dataframe_1["version"] \
                                   .astype("object")
processed_dataframe_1["level"] = processed_dataframe_1["level"] \
                                   .astype("UInt16")
processed_dataframe_1.dtypes
```

```python
                                            # Создание таблицы user_payments
```

```python
# построение таблицы с данными о платежах пользователей

user_payments = processed_dataframe_1.query("event_name == 'purchase'") \
                                     .rename(columns = {"user_time": "purchase_time", "extra_1": "item", 
                                       "extra_2": "offer_id", "param_2": "price"}) \
                                     .drop(['country', 'param_1', "event_name"], axis=1) \
                                     .reset_index(drop = True)

user_payments.insert(5, "purchase_number",  user_payments.groupby('user_id').cumcount()+1)
```

```python
# смотрим на фрейм

user_payments.head(10)
```

```python
user_payments.dtypes
```

```python
user_payments["purchase_number"] = user_payments["purchase_number"].astype("UInt16")
user_payments.dtypes
```

```python
# экспортируем фрейм данных в эксель-файл

user_payments.to_excel(r"D:\УНИВЕР\Product_analytics\Diploma\user_payments.xlsx", sheet_name ='user_payments', index = False)
```

```python
                                                # Создание таблицы user_info
```

```python
# делаем копию фрейма для дальнейших расчётов

copy_processed_dataframe_1 = processed_dataframe_1.copy(deep = True)

# проиндексировали фрейм "copy_processed_dataframe_1" колонкой "user_id"

copy_processed_dataframe_1.set_index("user_id", inplace = True)

# создание фрейма с нужной колонкой ('dates') со списком всех дат юзера за все сессии в игре

list_dates = copy_processed_dataframe_1.groupby('user_id', as_index=False) \
                                       ['user_time'].agg({'user_time':(lambda x: list(x))})

# создание фрейма с нужной колонкой ('total_revenue')

total_revenue = user_payments.groupby("user_id").price.sum() \
                             .reset_index(drop=False) \
                             .rename(columns={"price":"total_revenue"})

# создание фрейма с нужной колонкой ('last_time')

last_time = copy_processed_dataframe_1.groupby(level=0).last() \
                                      ["user_time"].reset_index()

# создание фрейма с нужной колонкой ('last_version')

last_version = copy_processed_dataframe_1.groupby(level=0).last() \
                                      ["version"].reset_index()

# создание фрейма с нужной колонкой ('max_level')

max_level = copy_processed_dataframe_1.groupby(level=0).last() \
                                      ["level"].reset_index()

# создание фрейма с нужной колонкой ('amount_gold')

amount_gold = copy_processed_dataframe_1.query("extra_1 == 'gold'") \
                                        .groupby(level=0).last() \
                                        .rename(columns={"param_2": "amount_gold"})

# создание фрейма с нужной колонкой ('amount_bronze_cup')

amount_bronze_cup = copy_processed_dataframe_1.query("extra_1 == 'bronze_cup'") \
                                        .groupby(level=0).last() \
                                        .rename(columns={"param_2": "amount_bronze_cup"})

# создание фрейма с нужной колонкой ('amount_silver_cup')

amount_silver_cup = copy_processed_dataframe_1.query("extra_1 == 'silver_cup'") \
                                        .groupby(level=0).last() \
                                        .rename(columns={"param_2": "amount_silver_cup"})

# создание фрейма с нужной колонкой ('amount_gold_cup')

amount_gold_cup = copy_processed_dataframe_1.query("extra_1 == 'gold_cup'") \
                                        .groupby(level=0).last() \
                                        .rename(columns={"param_2": "amount_gold_cup"})
```

```python
# построение таблицы с данными о пользователях

user_info = processed_dataframe_1.query("event_name == 'app_install_start'") \
                                 .rename(columns = {"extra_1": "type_trafic", "user_time": "first_time", 
                                                    "version": "first_version"}) \
                                 .drop(['extra_2', "level", "param_2", "param_1", "event_name", "session_id"], axis = 1) \
                                 .reset_index(drop=True) \
                                 .merge(total_revenue, how='left', on='user_id') \
                                 .merge(amount_gold["amount_gold"].astype("UInt32"), how='left', on='user_id') \
                                 .merge(amount_bronze_cup["amount_bronze_cup"].astype("UInt16"), how='left', on='user_id') \
                                 .merge(amount_silver_cup["amount_silver_cup"].astype("UInt16"), how='left', on='user_id') \
                                 .merge(amount_gold_cup["amount_gold_cup"].astype("UInt16"), how='left', on='user_id')

user_info.insert(4, "last_time", last_time["user_time"])
user_info.insert(5, "dates", list_dates["user_time"])
user_info.insert(6, "payer", user_info["user_id"].isin(user_payments["user_id"]).astype("UInt8"))
user_info.insert(8, "last_version", last_version["version"].astype("object"))
user_info.insert(9, "max_level", max_level["level"].astype("UInt16"))

# избавляемся от пустых значений (NaN), заменяя их на 0

user_info = user_info.fillna(0)

# смотрим на фрейм

user_info
```

```python
# смотрим тип данных всех колонок

user_info.dtypes
```

```python
# изменяем на нужный тип данных и смотрим на изменения

user_info["user_id"] = user_info["user_id"].astype("object")
user_info["first_version"] = user_info["first_version"].astype("object")
user_info["last_version"] = user_info["last_version"].astype("object")
user_info.dtypes
```

```python
                                            # Создание таблицы user_session
```

```python
# делаем копию фрейма для дальнейших расчётов

copy_processed_dataframe_2 = processed_dataframe_1.copy(deep = True)

# создание фрейма с нужной колонкой ('version') для таблицы 'user_session'

version_user_session = copy_processed_dataframe_2[["session_id", "version"]].groupby("session_id").last() \
                                                 .reset_index(drop=False)

# создание фрейма с нужной колонкой ('time_finish_session') для таблицы 'user_session'

time_finish_session = copy_processed_dataframe_2[["session_id", "user_time"]].groupby("session_id").last() \
                                                .rename(columns={"user_time":"time_finish_session"}) \
                                                .reset_index(drop=False)

# создание фрейма с нужной колонкой ('amount_get_level') для таблицы 'user_session'

amount_get_level = copy_processed_dataframe_2[["session_id", "level"]].groupby("session_id").last() \
                                                .rename(columns={"level":"amount_get_level"}) \
                                                .reset_index(drop=False)

# создание фрейма с нужной колонкой ('amount_get_gold') для таблицы 'user_session'

amount_get_gold = copy_processed_dataframe_2[["session_id", "extra_1", "param_2"]].query("extra_1 == 'gold'") \
                                        .groupby("session_id").last() \
                                        .rename(columns={"param_2": "amount_get_gold"}) \
                                        .reset_index(drop=False) \
                                        .drop(columns={"extra_1"}, axis=1)

# создание фрейма с нужной колонкой ('amount_get_bronze_cup') для таблицы 'user_session'

amount_get_bronze_cup = copy_processed_dataframe_2[["session_id", "extra_1", "param_2"]].query("extra_1 == 'bronze_cup'") \
                                        .groupby("session_id").last() \
                                        .rename(columns={"param_2": "amount_get_bronze_cup"}) \
                                        .reset_index(drop=False) \
                                        .drop(columns={"extra_1"}, axis=1)

# создание фрейма с нужной колонкой ('amount_get_silver_cup') для таблицы 'user_session'

amount_get_silver_cup = copy_processed_dataframe_2[["session_id", "extra_1", "param_2"]].query("extra_1 == 'silver_cup'") \
                                        .groupby("session_id").last() \
                                        .rename(columns={"param_2": "amount_get_silver_cup"}) \
                                        .reset_index(drop=False)\
                                        .drop(columns={"extra_1"}, axis=1)

# создание фрейма с нужной колонкой ('amount_get_gold_cup') для таблицы 'user_session'

amount_get_gold_cup = copy_processed_dataframe_2[["session_id", "extra_1", "param_2"]].query("extra_1 == 'gold_cup'") \
                                        .groupby("session_id").last() \
                                        .rename(columns={"param_2": "amount_get_gold_cup"}) \
                                        .reset_index(drop=False)\
                                        .drop(columns={"extra_1"}, axis=1)
```

```python
# построение таблицы

user_session = processed_dataframe_1.query("event_name == 'loading_finish'") \
                                    .rename(columns={"user_time": "time_start_session"}) \
                                    .drop(["version", "country", "level","event_name", \
                                           "param_1", "param_2", "extra_1", "extra_2"], axis=1) \
                                    .merge(time_finish_session, how ='inner', on ='session_id') \
                                    .merge(version_user_session, how ='inner', on ='session_id') \
                                    .merge(amount_get_level, how ='inner', on ='session_id') \
                                    .merge(amount_get_gold,how ='left', on ='session_id') \
                                    .merge(amount_get_bronze_cup, how ='left', on ='session_id') \
                                    .merge(amount_get_silver_cup, how ='left', on ='session_id') \
                                    .merge(amount_get_gold_cup, how ='left', on ='session_id')


user_session.insert(3, "session_number",  user_session.groupby('user_id').cumcount()+1)
user_session.insert(6, "length_session",  (user_session["time_finish_session"] - \
                                           user_session["time_start_session"]).dt.seconds)

# избавляемся от пустых значений (NaN), заменяя их на 0

user_session = user_session.fillna(0)

# смотрим на фрейм

user_session
```

```python
user_session.dtypes
```

```python
user_session["user_id"] = user_session["user_id"].astype("object")
user_session["session_number"] = user_session["session_number"].astype("UInt16")
user_session["length_session"] = user_session["length_session"].astype("UInt32")
user_session["version"] = user_session["version"].astype("object")
user_session["amount_get_gold"] = user_session["amount_get_gold"].astype("UInt32")
user_session["amount_get_bronze_cup"] = user_session["amount_get_bronze_cup"].astype("UInt16")
user_session["amount_get_silver_cup"] = user_session["amount_get_silver_cup"].astype("UInt16")
user_session["amount_get_gold_cup"] = user_session["amount_get_gold_cup"].astype("UInt16")
user_session.dtypes
```
